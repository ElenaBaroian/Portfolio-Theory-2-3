```python
import pandas_datareader.data as web

import pandas as pd 

import datetime

start = datetime.datetime(2013, 1, 1)

end = datetime.datetime(2019, 1, 1)

JPM = web.DataReader("JPM", 'yahoo', start, end)
SP500= web.DataReader("SPY", 'yahoo', start, end)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

JPM['Adj Close'].plot(label='JPM',figsize=(16,6))
SP500['Adj Close'].plot(label='SP')

plt.legend();
```

```python
#Use Date as index
JPM.reset_index(inplace=True,drop=False)
JPM.set_index('Date',inplace=True,drop=False)

#Extract only the EOM prices
JPM_EOM= JPM.groupby(pd.Grouper(key='Date',freq='MS')).apply(lambda x: x[-1:])
JPM_EOM.reset_index(inplace=True,drop=True)
JPM_EOM.set_index('Date',inplace=True,drop=True)
JPM_EOM
```

```python
SP500.reset_index(inplace=True,drop=False)
SP500.set_index('Date',inplace=True,drop=False)

SP500_EOM= SP500.groupby(pd.Grouper(key='Date',freq='MS')).apply(lambda x: x[-1:])
SP500_EOM.reset_index(inplace=True,drop=True)
SP500_EOM.set_index('Date',inplace=True,drop=True)
SP500_EOM
```

```python
#Plot the closing price
import matplotlib.pyplot as plt
%matplotlib inline

JPM_EOM['Adj Close'].plot(label='JPM',figsize=(16,6))
SP500_EOM['Adj Close'].plot(label='SP')

plt.legend();
```

```python
JPM_EOM['returns'] = JPM_EOM['Adj Close'].pct_change(1)
JPM_EOM.dropna(inplace=True)
SP500_EOM['returns'] = SP500_EOM['Adj Close'].pct_change(1)
SP500_EOM.dropna(inplace=True)
```

```python
#Plot returns
JPM_EOM['returns'].plot(label='JPM return',figsize=(16,6))
SP500_EOM['returns'].plot(label='SP return') 

plt.legend();
```

```python
#Scatter plot
plt.scatter(JPM_EOM['returns'],SP500_EOM['returns'],alpha=0.5);
```
